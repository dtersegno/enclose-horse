A line of N squares might have the highest edge to area ratio. Each square contributes 2 (top and bottom) plus 2 extras for the sides.

surface area = 2*N + 2
volume = N

With a large number of squares, SA/V -> 2

With one square, SA/V = 4

For a square-shaped region of N^2 squares total, the square is N wide, and so has surface area of 4*N. The SA/V ratio is 4/N. This agrees with the line of N = 1 squares.

So SA/V can limit to zero, and upper bound of 4. Adding any adjacent squares will at best contribute 3 edges to 1 square area.
Counting edges seems like a viable option.
